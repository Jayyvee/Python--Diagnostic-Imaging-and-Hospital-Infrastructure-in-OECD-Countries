# import libraries

```python
#Import Libraries
import pandas as pd
import numpy as np
import os
# visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

# import dataframes

```python
#Identify path 
data_path= '../02 Data/'
vis_path= '../04 Analysis/'
```

```python
df_health=pd.read_pickle(os.path.join(data_path,'Prepared Data','health_cleaned.pkl'))
```

# Correlations

```python
# create a subet of numeric only df
numeric_df = df_health.select_dtypes(include=['float64', 'int64'])
correlation = numeric_df.corr()
```

```python
# correlation matrix with panda
correlation
```

## Heatmap

```python
# create a subplot with matplotlib
f,ax=plt.subplots(figsize=(10,10))
# create heatmap with seaborn with the subplot of variables 
heat_map2=sns.heatmap(correlation.corr(),annot=True,ax=ax)

plt.title('Correlation Matrix', fontsize=14) # add title
```

## comment:
From the heatmap above we can notice strong correlations between MRI units and hospital bed. Indicating that care  is being outsouced to get an MRI vs having to admit a pt and take a bed for a simple scan.

```python
#Export 
heat_map2.figure.savefig(os.path.join(vis_path,'Visuals','seaborn heatmap.png'), bbox_inches='tight')
```

## heatmap using matplotlib

```python
# create correlation heatmap using matplotlib
heatmap=plt.matshow(correlation.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

heatmap= plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(correlation.corr(), fignum=f.number) # type of plot
plt.xticks(range(correlation.shape[1]), correlation.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(correlation.shape[1]), correlation.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
#Export matrix heatmap
heatmap.figure.savefig(os.path.join(vis_path,'Visuals','correlation matrix.png'), bbox_inches='tight')
```
